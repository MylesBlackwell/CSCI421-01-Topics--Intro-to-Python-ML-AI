```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.bar(categories, values)
plt.title("Category Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [20, 30, 25, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Simple Pie Chart")
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
data = { 'Product': ['A', 'B', 'A', 'B', 'C', 'A', 'C'], 'Region': ['North', 'South', 'South', 'North', 'North', 'South', 'South'], 'Sales': [100, 150, 200, 130, 90, 170, 50] }
df = pd.DataFrame(data)
df

pivot_table = pd.pivot_table(df, values='Sales', index='Product', columns='Region', aggfunc='sum')
print(pivot_table)

#what does pandas do and explain here
```
